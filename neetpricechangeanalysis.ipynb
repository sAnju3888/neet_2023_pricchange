```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt

import scienceplots
plt.style.use(['science','notebook','grid'])


#source is  https://en.wikipedia.org/wiki/National_Eligibility_cum_Entrance_Test_(Undergraduate)

registered_candidates = [1138890,1326725,1519375,1597435,1614777,1872343]
years = [2017,2018,2019,2020,2021,2022]
fees = [1400,1400,1400,1500,1500,1500]
ye = np.asarray(registered_candidates)

plt.plot(years,ye/10**5,marker='o',label="registration")
plt.xlabel("YEARS")
plt.ylabel("no of registered candidates(LAKHS)")
plt.legend()
plt.show()

plt.plot(years,fees)
plt.plot(2023, 1700, marker="o", markersize=10)
plt.xlabel("YEARS")
plt.ylabel("FEES")
plt.show()


# prediction of 2023 registered candidates
#linear regression 
def sse(p):

  p_m = np.mean(p)
  
  sse_1 = 0
  for y in p:
    sse_1 += (y-p_m)*(y-p_m)
  
  return sse_1

def co_var(p,q):
  i=0
  z=0
  for y in p:
       z+=( (y-np.mean(p)) * (q[i]-np.mean(q)) )
       
       i+=1
  return z

def coeff(p,q):
  b = []
  a2 = sse(p)
  cov = co_var(p,q)
  #b1
  b.append(cov/a2)
  #b0
  b.append(np.mean(q)-np.mean(p)*(b[0]))
  return b

def plot(p,q,pred):
  plt.scatter(years,ye/10**5,color="m")
  k = np.asarray(pred)
  plt.plot(years,k/10**5,color="red")
  plt.plot(2023, 1972538/10**5, marker="o", markersize=10)
  plt.ylabel("prediction candidates 2023")
  plt.xlabel("years")
  plt.show()


def main():
  p = years
  q = registered_candidates
  c = coeff(p,q)
  pred_values = []
  pred = "c[1]+c[0]*x"
  
  for x in p:
    pred_values.append(eval(pred))
  x=2023
  print(pred_values)
  print(eval(pred))
  plot(p,q,pred_values)



if __name__==main():
  main()

print("PROFIT GAINED BY INCREASING the neet fees", (1700*1972538) - (1500*1872343))
#54 cr is gain by changing 200 ruppess
print("To avoid our error accuracy lets take 50%  27 cr gained ")
```
